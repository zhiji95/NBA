```python
import numpy as np 
import pandas as pd
```

```python
df_event = pd.read_csv('NBA Hackathon - Event Codes.csv')
df_pbyp = pd.read_csv('NBA Hackathon - Play by Play Data Sample (50 Games).csv')
df_lineup = pd.read_csv('NBA Hackathon - Game Lineup Data.csv')
```

```python
# def pm_calculator(df1, df2, df3, person_id, game_id):

def get_event(event_code):
    codes = df_event['Event_Msg_Type']
    for i in range(len(codes)):
        if codes[i]  == event_code:
            return df_event['Event_Msg_Type_Description'][i]
    print ('The event code number does not exist')
    return None
def get_action(event_code):
    codes = df_event['Action_Type']
    for i in range(len(codes)):
        if codes[i]  == event_code:
            return df_event['Action_Type_Description'][i]
    print ('The action code number does not exist')
    return None
def get_teams(person_id):
#     print(list(df_lineup[df_lineup['Person_id'] == person_id]['Team_id'])[0])
    return list(df_lineup[df_lineup['Person_id'] == person_id]['Team_id'])[0],list(df_lineup[df_lineup['Person_id'] != person_id]['Team_id'])[0]

def get_pm(game_id, person_id, df,verbose = False):
# df is play by play data
    my_team, enemy = get_teams(person_id)
    plus_points, minus_points, tot_points = 0,0,0
    this_game = df[df['Game_id'] == game_id]
    for ped in np.r_[1,2,3,4]:
        this_period = this_game[this_game['Period'] == ped]
        zaichang = is_starters(person_id,ped,game_id)
        for i in range(len(this_period.index)):
            cur_team = list(this_period['Team_id'])[i]
            h = heuristic(cur_team, my_team)
            huanxia = list(this_period['Event_Msg_Type'])[i] == 8 and list(this_period['Person1'])[i] == person_id
            huanshang = list(this_period['Event_Msg_Type'])[i] == 8 and list(this_period['Person2'])[i] == person_id
            if (verbose):
                
                print('huanxia: ', huanxia, ',huanshang: ' ,huanshang, ',index : ',i, ',is starters: ',is_starters(person_id,ped,game_id))
            if (huanxia):
                zaichang = False
                j = i+1
                while(list(this_period['Event_Msg_Type'])[j] == 3):
                    plus_points += list(this_period['Option1'])[j]
                    j += 1
            if huanshang:
                zaichang = True
                j = i+1
                while(list(this_period['Event_Msg_Type'])[j] == 3):
                    minus_points += list(this_period['Option1'])[j]
                    j += 1
            if (zaichang):
    
#                 cur_team = list(this_period['Team_id'])[i]
#                 h = heuristic(cur_team, my_team)
                tot_points += list(this_period['Option1'])[i] * h
#                 if (huanxia):
#                     j = i
#                     while(list(this_period['Event_Msg_Type'])[j] == 3):
#                         plus_points += list(this_period['Option1'])[j]* h
#                         j += 1
#                 if (huanshang):
#                     j = i
#                     while(list(this_period['Event_Msg_Type'])[j] == 3):
#                         minus_points += list(this_period['Option1'])[j]* h
#                         j += 1
            
    return tot_points - minus_points + plus_points

def heuristic(cur_team, my_team):
    if cur_team == my_team:
        return 1
    return -1
    
def is_starters(person_id,period, game_id):
    starters = df_lineup[df_lineup['Period'] == period][df_lineup['Game_id'] == game_id]['Person_id']
    return person_id in list(starters)
```

```python
def get_all_players(game_id):
    result = []  
    this_game = df_lineup[df_lineup['Game_id'] == game_id]
    for i in list(this_game['Person_id']):
        if not (i in result):
            result.append(i)
    return result
def get_all_game_id():
    result = []
    for i in list(df_lineup['Game_id']):
        if not(i in result):
            result.append(i)
    return result
```

```python
df = pd.DataFrame()
gids, pids, pms = [],[],[]
for gid in get_all_game_id():
    for pid in get_all_players(gid):
        gids.append(gid)
        pids.append(pid)
        pms.append(get_pm(gid, pid,df_pbyp))
df['Game_id'] = gids
df['Person_id'] = pids
df['Plus/Minus'] = pms
df
        
```
